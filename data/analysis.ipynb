```python
import pandas as pd
```

```python
data = pd.read_csv('scan2.csv')
data['speedup'] = data.apply(lambda row: row['scan_time']/row['parallel_scan_time'], axis=1)
data['efficiency'] = data.apply(lambda row: row['speedup']/row['num_threads'], axis=1)
data = data.set_index('num_threads')
data
```

```python
data[['speedup']].plot()
```

```python
data[['efficiency']].plot()
```
